```
#importing the matplotlib library
import matplotlib.pyplot as plt
```

```
# ploting a data using a list
views = [300,500,320,800,460,320]
days=['Mon','Tue','Wed','Thurs','Fri','Sat']
plt.plot(days,views, label ='youtube views', color = 'Black', marker = 'o', markerfacecolor ='r', linestyle='--', linewidth = 3)
plt.xlabel('Days of the week')
plt.ylabel('views')
plt.legend(loc= 'upper left')
plt.title('Youtube view for the week')
plt.show()
```

```
# looking at the relationship between various food ordered at the canteen 
moi_moi= [40,60,70,80,90]    
rice=[30,50,40,70,30]
bread=[30,40,50,20,80]
beans=[60,45,55,25,85]
days=['mon','tue','wed','thurs','fri']
plt.plot(days,moi_moi, label= 'moi_moi',color = 'black', marker ='o', markerfacecolor= 'black', linestyle ='--' )
plt.plot(days,rice, label= 'rice',color = 'blue', marker ='o', markerfacecolor= 'black', linestyle ='--' )
plt.plot(days,beans, label= 'beans',color = 'green', marker ='o', markerfacecolor= 'black', linestyle ='--' )
plt.plot(days,bread, label= 'bread',color = 'yellow', marker ='o', markerfacecolor= 'black', linestyle ='--' )
plt.legend()
plt.grid(True, linewidth= 1, color = 'black', linestyle = '--')
plt.title('Food Distribution For The Week')
plt.xlabel('daysof the week')
plt.ylabel('number  of plates sold')
plt.show()
# using plt.xlim to limit the data showed on the x axis 
# using the plt.ylim to limit the data showed on the y axis 
# you can also use it to remove outliers 
# saving an image using my code 
# Note: the parameters are very key you use 
# plt.savefig(location, dpi= 'resoluton' color='true')
```
